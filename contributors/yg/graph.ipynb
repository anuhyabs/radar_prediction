```python
import pandas as pd

df = pd.read_csv ('2.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = df['Grain size']
y = df['zeta']
x_new = np.linspace(x.min(), x.max(),500)
f = interp1d(x, y, kind='quadratic')
y_smooth=f(x_new)
plt.plot (x_new,y_smooth)
plt.scatter (x, y)
plt.xlabel('Grain size(mm)')
plt.ylabel('Zeta(1/m)')
plt.show()
```
